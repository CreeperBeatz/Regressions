# Decision Tree Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Decision Tree Regression model on the whole dataset

```
from sklearn import tree

# random_state = 0, so we get the same result every time
clf = tree.DecisionTreeRegressor(random_state=0) 
clf.fit(X, y)
```

## Predicting a new result

```
clf.predict([[6.5]])
```

## Visualising the Decision Tree Regression results (low resolution)

```
plt.scatter(X,y, color = 'red')
plt.plot(X, clf.predict(X), color = 'blue')
plt.title('Truth or bluff (Decision tree regression)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

## Visualising the Decision Tree (High resolution)

```
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X,y, color = 'red')
plt.plot(X_grid, clf.predict(X_grid), color = 'blue')
plt.title('Truth or bluff (Decision tree regression)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```
